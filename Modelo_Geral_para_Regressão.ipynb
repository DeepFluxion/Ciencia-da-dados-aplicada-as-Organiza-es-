### O que é análise de regressão?

Redman oferece este exemplo de cenário: suponha que você seja um gerente de vendas tentando prever os números do próximo mês. Você sabe que dezenas, talvez até centenas de fatores, desde o clima até a promoção de um concorrente e o boato de um modelo novo e aprimorado podem afetar o número. Talvez as pessoas da sua organização tenham até uma teoria sobre o que terá maior efeito nas vendas. "Confie em mim. Quanto mais chuva temos, mais vendemos. ” "Seis semanas após a promoção do concorrente, as vendas aumentam."

A análise de regressão é uma maneira de classificar matematicamente quais dessas variáveis ​​realmente têm impacto. Responde às perguntas: Quais fatores são mais importantes? O que podemos ignorar? Como esses fatores interagem entre si? E, talvez o mais importante, até que ponto estamos certos sobre todos esses fatores?

Na análise de regressão, esses fatores são chamados de variáveis. Você tem sua variável dependente - o principal fator que você está tentando entender ou prever. No exemplo de Redman acima, a variável dependente são as vendas mensais. E então você tem suas variáveis independentes - os fatores que você suspeita têm um impacto sobre sua variável dependente.

### Como funciona?

Para realizar uma análise de regressão, você reúne os dados nas variáveis em questão. (Lembrete: você provavelmente não precisa fazer isso sozinho, mas é útil entender o processo que seu colega analista de dados usa.) Você usa todos os seus números de vendas mensais nos últimos três anos, por exemplo, nos últimos três anos e em qualquer dado sobre as variáveis independentes nas quais você está interessado. Portanto, neste caso, digamos que você descubra também a precipitação média mensal nos últimos três anos. Em seguida, você plota todas essas informações em um gráfico semelhante a este:

![Como gráfico incial regressão](https://hbr.org/resources/images/article_assets/2015/10/W20151029_GALLO_RELATIONSHIPVARIABLES.png)

O eixo y é a quantidade de vendas (a variável dependente, a coisa em que você está interessado, está sempre no eixo y) e o eixo x é a precipitação total. Cada ponto azul representa os dados de um mês - quanto choveu naquele mês e quantas vendas você fez no mesmo mês.


Olhando para esses dados, você provavelmente percebe que as vendas são maiores nos dias em que chove muito. É interessante saber, mas em quanto? Se chover 3 polegadas, você sabe quanto vai vender? E se chover 10 cm?

Agora imagine desenhar uma linha no gráfico acima, que percorre aproximadamente o meio de todos os pontos de dados. Essa linha o ajudará a responder, com certo grau de certeza, quanto você normalmente vende quando chove uma certa quantidade.

![Como gráfico incial regressão](https://hbr.org/resources/images/article_assets/2015/10/W20151029_GALLO_REGRESSIONMODEL.png)

Isso é chamado de linha de regressão e é desenhado (usando um programa de estatísticas como SPSS, STATA ou mesmo Excel) para mostrar a linha que melhor se ajusta aos dados. Em outras palavras, explica Redman, "a linha vermelha é a melhor explicação do relacionamento entre a variável independente e a variável dependente".

Além de desenhar a linha, seu programa de estatísticas também gera uma fórmula que explica a inclinação da linha e se parece com isso:

## Y = 200 + 5*X + erro

Ignore o termo de erro por enquanto. Refere-se ao fato de que a regressão não é perfeitamente precisa. Apenas se concentre no modelo:

## Y = 200 + 5*X

O que essa fórmula está dizendo é que, se não houver um "x", então Y = 200. Então, historicamente, quando não choveu, você fez uma média de 200 vendas e pode esperar fazer o mesmo daqui para frente. assumindo que outras variáveis ​​permaneçam iguais. E no passado, para cada centímetro adicional de chuva, você fazia em média mais cinco vendas. "Para cada incremento que x sobe um, y sobe cinco", diz Redman.

Agora vamos voltar ao termo do erro. Você pode ficar tentado a dizer que a chuva tem um grande impacto nas vendas se a cada centímetro você conseguir mais cinco vendas, mas se essa variável vale a pena, sua atenção dependerá do termo do erro. Uma linha de regressão sempre tem um termo de erro porque, na vida real, variáveis ​​independentes nunca são preditores perfeitos das variáveis ​​dependentes. Em vez disso, a linha é uma estimativa com base nos dados disponíveis. Portanto, o termo do erro informa o quanto você pode ter certeza sobre a fórmula. Quanto maior, menor a certeza da linha de regressão.

O exemplo acima usa apenas uma variável para prever o fator de interesse - nesse caso, chuva para prever vendas. Normalmente, você inicia uma análise de regressão que deseja entender o impacto de várias variáveis ​​independentes. Portanto, você pode incluir não apenas chuva, mas também dados sobre a promoção de um concorrente. "Você continua fazendo isso até que o termo do erro seja muito pequeno", diz Redman. "Você está tentando obter a linha que melhor se ajusta aos seus dados." Embora possa haver perigos em tentar incluir muitas variáveis ​​em uma análise de regressão, analistas qualificados podem minimizar esses riscos. E considerar o impacto de várias variáveis ​​ao mesmo tempo é uma das maiores vantagens da regressão.

# 1. Regressão Linear

Modelos de regressão linear são um bom ponto de partida para tarefas de regressão. Esses modelos são populares porque podem ser ajustados muito rapidamente e são muito interpretáveis. Você provavelmente está familiarizado com a forma mais simples de um modelo de regressão linear (ou seja, ajustando uma linha reta aos dados), mas esses modelos podem ser estendidos para modelar comportamentos de dados mais complicados.

Nesta aula, começaremos com uma rápida explicação intuitiva da matemática por trás desse problema conhecido, antes de ver como antes passar para ver como os modelos lineares podem ser generalizados para explicar padrões mais complicados nos dados.

Começamos com as importações padrão:

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

## Regressão linear simples

Começaremos com a regressão linear mais familiar, um ajuste linear aos dados.
Um ajuste de linha reta é um modelo do formulário
$$
y = ax + b
$$
onde $ a $ é conhecido como **inclinação (slope)** e $ b $ é conhecido como **intercepto (intercept)**.

Considere os seguintes dados, que estão espalhados por uma linha com uma inclinação de 2 e uma interceptação de -5:

```python
rng = np.random.RandomState(1)
x = 10 * rng.rand(50)
y = 2 * x - 5 + rng.randn(50)
plt.scatter(x, y);
```

Podemos usar o estimador `` LinearRegression`` do Scikit-Learn para ajustar esses dados e construir a linha de melhor ajuste:

```python
from sklearn.linear_model import LinearRegression
linreg = LinearRegression(fit_intercept=True)

linreg.fit(x[:, np.newaxis], y)

xfit = np.linspace(0, 10, 1000)
yfit = linreg.predict(xfit[:, np.newaxis])

plt.scatter(x, y, label='amostra')
plt.plot(xfit, yfit,'r--', label='valores ajustados');
plt.legend()
```

O y é o nosso valor a ser predito, vamos ver como os dado se sobrepõe

```python
plt.plot(y,'bo-', label='y-valor real');
plt.plot(linreg.predict(x[:, np.newaxis]),'r--',label='yfit-valor predito')
plt.legend()
```

A inclinação e o intercepto dos dados estão contidas nos parâmetros de ajuste do modelo, que no Scikit-Learn são sempre marcados por um sublinhado à direita. Aqui os parâmetros relevantes são ``coef_`` e ``intercept_``:

```python
print("Inclinação Modelo:    ", linreg.coef_[0])
print("intercepto Modelo:", linreg.intercept_)
```

# 2 Regressão Linear Multipla

o estimador `` Regressão linear`` é muito mais capaz que isso, no entanto - além de ajustes simples em linha reta, ele também pode lidar com modelos lineares multidimensionais da forma
$$
y = a_0 + a_1 x_1 + a_2 x_2 + \ cdots
$$
onde existem vários valores de $ x $.
Geometricamente, isso é semelhante ao ajuste de um plano a pontos em três dimensões ou ao ajuste de um hiperplano a pontos em dimensões mais altas.

A natureza multidimensional de tais regressões as torna mais difíceis de visualizar, mas podemos ver um desses ajustes em ação construindo alguns dados de exemplo, usando o operador de multiplicação de matrizes de NumPy:

```python
rng = np.random.RandomState(1)
X = 10 * rng.rand(100, 3)
y = 0.5 + np.dot(X, [1.5, -2., 1.])

linreg.fit(X, y)
print("intercepto Modelo:", linreg.intercept_)
print("Inclinação Modelo:", linreg.coef_)
```

o estimador `` Regressão linear`` agora possibilita definir os termos $a_0$ $a_1$ e $a_2$ e o $intecepto$ da equação:
$$
y = a_0 + a_1 x_1 + a_2 x_2 + \ cdots
$$

Obtemos então uma equação com **coeficientes numéricos** que permite calcular valores y conhecendo $x_1$ e $x_2$

$$
y = 1.5 - 2.0 x_1 + 1.0 x_2
$$

## DataFrame dos Dados

```python
reg_mult = pd.DataFrame([X[:,0],X[:,1],y]).T
reg_mult.columns = ['X1','X2','y'];
reg_mult
```

Aqui, os dados y são construídos a partir de três valores aleatórios x, e a regressão linear recupera os coeficientes usados para construir os dados.

Dessa maneira, podemos usar o único estimador LinearRegression para ajustar linhas, planos ou hiperplanos aos nossos dados. Ainda parece que essa abordagem seria limitada a relacionamentos estritamente lineares entre variáveis, mas acontece que podemos relaxar isso também.

```python
reg_mult.boxplot()
```

## 3. Regressão Não Linear

REGRESSÃO
Regressão é geralmente definida como a determinação de relacionamentos entre duas ou mais variáveis. Por exemplo, considere que você deve prever a renda de uma pessoa, com base nos dados de entrada X fornecidos.

Aqui, a variável de destino significa a variável desconhecida com a qual nos preocupamos em prever e contínua significa que não há lacunas (descontinuidades) no valor que Y pode assumir.

Prever renda é um problema clássico de regressão. Seus dados de entrada devem ter todas as informações (conhecidas como recursos) sobre o indivíduo que podem prever renda, como horário de trabalho, experiência educacional, cargo e local.

![Visão geral de regressõa no scikit learn](https://miro.medium.com/max/2000/1*Ue8DVc9CcHrIqn9dhko0Tg.png)

### Métricas (Geral)

Não vamos abordar todas neste tutorial tabalharemos co $R^2$

A função r2_score calcula o coeficiente de determinação, geralmente indicado como R².

Representa a proporção de variância (de y) que foi explicada pelas variáveis independentes no modelo. Ele fornece uma indicação da qualidade do ajuste e, portanto, uma medida de quão bem as amostras não vistas provavelmente serão previstas pelo modelo, por meio da proporção da variação explicada.

Como essa variação depende do conjunto de dados, R² pode não ser significativamente comparável entre diferentes conjuntos de dados. A melhor pontuação possível é 1,0 e pode ser negativa (porque o modelo pode ser arbitrariamente pior). Um modelo constante que sempre prediz o valor esperado de y, desconsiderando os recursos de entrada, obteria uma pontuação de R² de 0,0.

O scikit learn possui várias métricas de análise veja a figura

![Visão geral de regressõa no scikit learn](https://miro.medium.com/max/2000/1*ypc0PNKXsDlJR-1AaKnOeQ.png)



```python
rng = np.random.RandomState(42)
x = 10 * rng.rand(1000)

def senoide(x, sigma=0.8):
    fast_oscillation = np.sin(5 * x)
    slow_oscillation = np.sin(0.5 * x)
    noise = sigma * rng.randn(len(x))

    return slow_oscillation + fast_oscillation + noise

y = senoide(x)
plt.figure(figsize=(15,5))
plt.errorbar(x, y, 0.3, fmt='o');
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(x[:, None], y)

xfit = np.linspace(0, 10, 1000)
yfit = reg.predict(xfit[:, None])
ytrue = senoide(xfit, sigma=0)

plt.figure(figsize=(15,5))
plt.errorbar(x, y, 0.3, fmt='o', alpha=0.1, label='dados')
plt.plot(xfit, yfit, '-r', lw=2, label='modelo');
plt.plot(xfit, ytrue, '-k', alpha=0.5, label='senoide');
plt.text(5,4,'score= '+str(np.round(reg.score(xfit[:, None],ytrue),3)),fontsize=12)
plt.title('Linear Regressor', fontsize=18);
plt.legend();
plt.show()
```

```python
from sklearn.tree import DecisionTreeRegressor
reg = DecisionTreeRegressor(max_depth=3)
reg.fit(x[:, None], y)

xfit = np.linspace(0, 10, 1000)
yfit = reg.predict(xfit[:, None])
ytrue = senoide(xfit, sigma=0)

plt.figure(figsize=(15,5))
plt.errorbar(x, y, 0.3, fmt='o', alpha=0.1, label='dados')
plt.plot(xfit, yfit, '-r', lw=2, label='modelo');
plt.plot(xfit, ytrue, '-k', alpha=0.5, label='senoide');
plt.text(5,4,'score= '+str(np.round(reg.score(xfit[:, None],ytrue),3)),fontsize=12)
plt.title('Decision Tree Regressor', fontsize=18);
plt.legend();
plt.show()
```

```python
from sklearn.ensemble import RandomForestRegressor
reg = RandomForestRegressor(150)
reg.fit(x[:, None], y)

xfit = np.linspace(0, 10, 1000)
yfit = reg.predict(xfit[:, None])
ytrue = senoide(xfit, sigma=0)

plt.figure(figsize=(15,5))
plt.errorbar(x, y, 0.3, fmt='o', alpha=0.1, label='dados')
plt.plot(xfit, yfit, '-r', lw=2, label='modelo');
plt.plot(xfit, ytrue, '-k', alpha=0.5, label='senoide');
plt.text(5,4,'score= '+str(np.round(reg.score(xfit[:, None],ytrue),3)),fontsize=12)
plt.title('RandomForest Regressor', fontsize=18);
plt.legend();
plt.show()
```

```python
from sklearn.ensemble import AdaBoostRegressor
reg = AdaBoostRegressor()
reg.fit(x[:, None], y)

xfit = np.linspace(0, 10, 1000)
yfit = reg.predict(xfit[:, None])
ytrue = senoide(xfit, sigma=0)

plt.figure(figsize=(15,5))
plt.errorbar(x, y, 0.3, fmt='o', alpha=0.1, label='dados')
plt.plot(xfit, yfit, '-r', lw=2, label='modelo');
plt.plot(xfit, ytrue, '-k', alpha=0.5, label='senoide');
plt.text(5,4,'score= '+str(np.round(reg.score(xfit[:, None],ytrue),3)),fontsize=12)
plt.title('AdaBoost Regressor', fontsize=18);
plt.legend();
plt.show()
```

```python
from sklearn.ensemble import ExtraTreesRegressor
reg = ExtraTreesRegressor(100)
reg.fit(x[:, None], y)

xfit = np.linspace(0, 10, 1000)
yfit = reg.predict(xfit[:, None])
ytrue = senoide(xfit, sigma=0)

plt.figure(figsize=(15,5))
plt.errorbar(x, y, 0.3, fmt='o', alpha=0.1, label='dados')
plt.plot(xfit, yfit, '-r', lw=2, label='modelo');
plt.plot(xfit, ytrue, '-k', alpha=0.5, label='senoide');
plt.text(5,4,'score= '+str(np.round(reg.score(xfit[:, None],ytrue),3)),fontsize=12)
plt.title('ExtraTrees Regressor', fontsize=18);
plt.legend();
plt.show()
```
